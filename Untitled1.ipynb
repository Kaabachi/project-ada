```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
read = pd.read_csv("../data/food-inspections.csv")
```

```python
data = read.copy()
```

```python
data.describe()
```

```python
data.head()
```

```python
data = data.drop(columns=["Historical Wards 2003-2015","Zip Codes", "Community Areas", "Census Tracts", "Wards"])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=data.isnull().sum()* 100 /data.count(),y=data.isnull().sum().index)
plt.xlabel("Percentage")
plt.title("Percentage of missing value",size=20)
```

### Inspection Date

We observe that 35 % of the data for vialotion is missing. 

```python
data["Inspection Date"]
```

We transform the inspection date to several columns: Year, Month and Day 

```python
data['Inspection Date'] = data['Inspection Date'].map(lambda x: x.split('-'))
data['Year'] = data['Inspection Date'].map(lambda x: int(x[0]))
data['Month'] = data['Inspection Date'].map(lambda x: int(x[1]))
data['Day'] = data['Inspection Date'].map(lambda x: int(x[2].split('T')[0]))
data = data.drop(columns=['Inspection Date'])
```

```python
fig,ax=plt.subplots(2,2,figsize=(20,20))

x=data.Year.value_counts().index
y=data.Year.value_counts()
sns.barplot(x=x,y=y,ax=ax[0,0])
ax[0,0].set_title("The counts of inspection by year")
ax[0,0].set_ylabel('counts')
ax[0,0].set_xlabel('')

sns.scatterplot(x='Longitude',y='Latitude', hue='Year' ,data=data, ax=ax[0,1])

x=data.Month.value_counts().index
y=data.Month.value_counts()
sns.barplot(x=x,y=y,ax=ax[0,1])
ax[0,1].set_title("The counts of inspection by month")
ax[0,1].set_ylabel('counts')
ax[0,1].set_xlabel('')


x=data.Day.value_counts().index
y=data.Day.value_counts()
sns.barplot(x=x,y=y,ax=ax[1,0])
ax[1,0].set_title("The counts of inspection by day")
ax[1,0].set_ylabel('counts')
ax[1,0].set_xlabel('')

data.groupby(['Year','Month'])['Inspection ID'].agg('count').unstack('Year').plot(ax=ax[1,1])
ax[1,1].set_title("The counts of inspection for every month by year")
ax[1,1].set_ylabel('counts')
ax[1,1].set_xlabel('month')
```
