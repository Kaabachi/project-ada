An inspection can pass, pass with conditions or fail. 

    -Pass : Establishments were found to have no critical or serious violations (violation number 1-14 and 15-

### Inspection Type

```python
data.dropna(subset=['Inspection Type'], inplace=True)
```

There is only one null value for Inspection type, we drop it.

An inspection can be one of the following types: 
    
    -canvass, the mostcommon type of inspection performed at a frequency relative to the risk of the establishment;
    
    -consultation, when the inspection is done at the request of the owner prior to the opening of the establishment; 
    
    -complaint, when the inspection is done in response to a complaint against the establishment;
    
    -license, when the inspection is done as a requirement for the establishment to receive its license to operate;
    
    -suspect food poisoning, when the inspection is done in response to one or more persons claiming to have gotten ill as a result of eating at the establishment (a specific type of complaint-based inspection);

    -task-force inspection, when an inspection of a bar or tavern is done.

Re-inspections can occur for most types of these inspections and are indicated as such. 


```python
data["Inspection Type"].isnull().sum()
```

```python
data["Inspection Type"].unique()
```

We observe that we have several inspection's types representing the same thing and thus we can regroup them.

```python
data["Inspection Type exact"] = data["Inspection Type"].copy()
```

```python
data["Inspection Type"] = data["Inspection Type"].map(lambda x: x.lower())
data["Inspection Type"] = data["Inspection Type"].map(lambda x: re.sub('re-', 're', x))
data["Inspection Type"] = data["Inspection Type"].map(lambda x:'reinspection' if ('reinspection' in x ) else x)
data["Inspection Type"] = data["Inspection Type"].map(lambda x:'canvass' if ('canvas' in x ) else x)
data["Inspection Type"] = data["Inspection Type"].map(lambda x:'complaint' if ('complaint' in x ) else x)
data["Inspection Type"] = data["Inspection Type"].map(lambda x:'license' if ('license' in x ) else x)
data["Inspection Type"] = data["Inspection Type"].map(lambda x:'task force' if ('task' in x ) else x)
```

Staphylococcal food poisoning (SFP) is one of the most common food-borne diseases worldwide.

The number of inspections for food poisoning is interesting and give us some informations.
Indeed, from 2010 to 2014 we observe that while the total number of food inspections increases by a factor of 1.2, 
there is a net decrease (by a factor of 5) of the food poisoning inspections. 

So what did change, several hypohtesis are possible : 
    
    -H0: The detection algorithm improved so that from the number of inspections needed is smaller
    -H1: The inspection had a direct effect and the bad restaurants were closed 

Let's try to test these hypohtesis:

```python
data['Fail ?'] = data['Results'] == "Fail"
data['Pass ?'] = data['Results'] == "Pass"
data['Pass w/ Conditions ?'] =  data['Results'] == "Pass w/ Conditions"

fig = plt.figure(figsize=(10,5))

ax = fig.add_subplot(131) 
ax2 = fig.add_subplot(132) 
ax3 = fig.add_subplot(133) 

failures = data.groupby("Inspection Type").sum()['Fail ?']
successes = data.groupby("Inspection Type").sum()['Pass ?']
condtions = data.groupby("Inspection Type").sum()['Pass w/ Conditions ?']

group1 = ['canvass', 'complaint', 'license', 'reinspection']
group2 = ['consultation' , 'other' , 'recent inspection', 'tag removal', 'task force']

failures[group1].plot(kind ='bar',ax=ax, color='r',position=0,width=0.3,legend=True)
condtions[group1].plot(kind ='bar', ax=ax, color='y',position=1,width=0.3,legend=True)
successes[group1].plot(kind ='bar', ax=ax, color='g',position=2,width=0.3,legend=True)

failures[group2].plot(kind ='bar',ax=ax2, color='r',position=0,width=0.3,legend=True)
condtions[group2].plot(kind ='bar', ax=ax2, color='y',position=1,width=0.3,legend=True)
successes[group2].plot(kind ='bar', ax=ax2, color='g',position=2,width=0.3,legend=True)

pd.Series(failures['food poisoning']).plot(kind ='bar',ax=ax3, color='r',position=0,width=0.3,legend=True)
pd.Series(condtions['food poisoning']).plot(kind ='bar', ax=ax3, color='y',position=1,width=0.3,legend=True)
pd.Series(successes['food poisoning']).plot(kind ='bar', ax=ax3, color='g',position=2,width=0.3,legend=True)
```
