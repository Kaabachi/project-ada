```python
#Import dataset from previously obtained CSV file
restaurants_food_poisoning = pd.read_csv("data/sample_food_poisoning.csv").drop(columns = ['Unnamed: 0'])

#Extract rating from reviewRating
restaurants_food_poisoning["trueRating"]=restaurants_food_poisoning["reviewRating"].str.extract('(\d+)')

restaurants_food_poisoning = restaurants_food_poisoning.astype({'trueRating': 'int32'})
restaurants_food_poisoning.dtypes
```

```python
restaurants_food_poisoning
```

```python
# Distribution of review ratings for these restaurants
restaurants_food_poisoning.groupby("trueRating").count().plot(y='alias',kind='bar')
```

```python
def preprocess(x):
    x = re.sub('[^a-z\s]', '', x.lower())                  # get rid of noise
    x = [w for w in x.split() if w not in set(stopwords)]  # remove stopwords
    return ' '.join(x) # then join the text again
```

```python
#nltk.download('stopwords')
i = nltk.corpus.stopwords.words('english')
j = list(string.punctuation)
stopwords = set(i).union(j)
```

```python
restaurants_food_poisoning["description_filtered"]= restaurants_food_poisoning.apply(lambda row : preprocess(row["description"]) , axis = 1)
```

```python
restaurants_food_poisoning
```

```python
wc = WordCloud(width=1600, height=800, random_state=1, max_words=20000)
wc.generate(' '.join(restaurants_food_poisoning['description_filtered']))

# declare our figure 
plt.figure(figsize=(20,10), facecolor='k')
# add title to the graph
plt.title("Reviews for restaurants with food poisoning", fontsize=40,color='white')
plt.imshow(wc)
plt.axis('off')
plt.tight_layout(pad=10)
```
